```python
import numpy as np
```

```python
#PART 1
#Getting 64*64 Matrix
matrix = np.random.rand(64,64)*255
```

```python
matrix
```

```python
#Slicing 64*64 matrix into 64 such 64*1 
arr = np.empty(64)
for i in range(0,64):
    for j in range(0,64):
        arr[j] = matrix[j][i]
    print("Vector ",i+1," is:\n")
    print(arr,"\n")
    sum,eq,mean,variance = 0,0,0,0
    # Calculating Mean 
    for p in range(64):
        sum=sum+arr[p]
    mean=sum/64
    print("Mean for this Vector is: ",mean)
    #Calculating Variance
    for q in range(64):
        eq=eq+float((float(arr[q])-float(mean))**2)
    variance=float(eq/64)
    print("Variance for this Vector is: ",variance,"\n")
```

```python
#COVARIANCE

#Creating empty Array
arr_1=np.empty(64)
arr_2=np.empty(64)
#Covariance of Vector 'i' and Vector 'j'

#Taking input of first vector value in range 0,64
i =-1
while not int(i) in range(64):
    i = int(input("Enter the first Vector: "))
for j in range(64):
    arr_1[j]=matrix[j][i]
m1=np.mean(arr_1)

#Taking input of second vector value in range 0,64
i =-1
while not int(i) in range(64):
    i = int(input("Enter the second Vector: "))
for j in range(64):
    arr_2[j]=matrix[j][i]
m2=np.mean(arr_2)

# Actual Calculation
sum_cov=0
for y in range(64):
    sum_cov+=(arr_1[y]-m1)*(arr_2[y]-m2)
cov_1=sum_cov/64
print("Covariance between above entered vectors is: ",cov_1)
```
